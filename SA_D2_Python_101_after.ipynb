# Classes

* Class vs Class Instance
* methods and attributes
* __init__(self)
* __repr__(self)/__str__(self)
* Example: Car class
  * attributes: make, model, year, _odometer (private)
  * methods:
    * __init__(self)
    * desribe_car(self)
    * update_odometer(self, val) /no rollback/
    * increment_odomoter(self, val) /check positive val else error/
    * describe_care(self)
    * __str__(self)
* Static methods
* Polymorphism

```python
class Car:

  def __init__(self, make, model, year):
    self.make = make
    self.model = model
    self.year = year
    self._odometer = 0

  def __str__(self):
    return self.make + ": " + self.model

  def describe_car(self):
    print("This car is ", self.make, ":", self.model, "made in", self.year, ".Kms", self._odometer)

  def update_odometer(self, value):
    if value > self._odometer:
      self._odometer = value
    else:
      raise Exception("Cannot rollback odometer")

  @staticmethod
  def make_car_sound():
    print("Vruum")


x = Car("Škoda", "Yetti", 2005)
x.describe_car()
Car.make_car_sound()

#x.describe_car()
#x.update_odometer(150000)
#x.describe_car()
#y = Car("VW", "Arteon", 2010)
#y.describe_car()
```

### <font color="orange">TASKS (15 min) </font>
Create a class Point2D which will take two attributes (self.x, self.y). Create method for __init__, __str__ -> "Point(<x>,<y>)", and distance of two points - the method will have one argument and will accept another Point2D.

```python
import math
class Point2D:

  def __init__(self, x, y):
    self.x = x
    self.y = y

  def __str__(self):
    return "Point(" + str(self.x) + "," + str(self.y) + ")"

  def distance(self, point):
    return math.sqrt((self.x - point.x) ** 2 + (self.y - point.y) ** 2)

a = Point2D(10, 5)
b = Point2D(25, 0)

print(a)
print("Distance a-b:", a.distance(b))
```

```python
# If you use Google Colab, use this
rand_word = sample(words.words(), 1)[0]
#print(rand_word)
rand_word = "Airplane"
lives = 3

# TODO:
def print_current_state(word, letters_guessed, lives):
  res = ""
  for char in word:
    if char.lower() in letters_guessed:
      res += char + " "
    else:
      res += "_ "
  res += " | " + str(lives) + " left"
  print(res)
# Game Start
# Initialize set letters_guessed, letters_word
letters_guessed = set()
letters_word = set(rand_word.lower())
print(f"Game start: You have {lives} lives. Word is {len(rand_word)} long.")
while lives > 0:
  letter = input("Guess a letter:")
  if letter in letters_guessed:
    print("You already guessed that, what are you, fish? o.O")
    continue
  else:
    # if yes, add it to letters_guessed
    letters_guessed.add(letter)
      # if not - decrease lives
    if letter not in letters_word:
      lives -= 1
      if lives == 0:
        print("You lost!!")
        break
  # print the current state (This is the hardest part) of the game -> a _ _ _ _ a _ _ | X lives left
  print_current_state(rand_word, letters_guessed, lives)
  if len(letters_guessed.intersection(letters_word)) == len(letters_word):
    print("You won!!")
    break
```
